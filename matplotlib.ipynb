```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#global api
x = np.arange(-10, 11)
plt.figure(figsize=(12,6))
plt.title('New Plot')

plt.subplot(1,2,1) #row, col, selected subplot
plt.plot(x, x**2)
plt.plot([0,0,0], [-10,0,100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('x')
plt.ylabel('x squared')

plt.subplot(1,2,2) #row, col, selected subplot
plt.plot(x, -(x**2))
plt.plot([-10,0,10],[-50,-50,-50])
plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('x')
plt.ylabel('x squared')
```

```python
#oop api
fig, axes = plt.subplots(figsize=(12,6))
```

```python
axes.plot(
    x, (x**2), color='red',linewidth=2,
    marker='o', label='X^2')

axes.plot(
    x, -(x**2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X squared')

axes.set_title('My New Plot')

axes.legend()

fig
```

```python
values = np.random.randn(1000)
plt.subplots(figsize=(12,6))
plt.hist(values, bins=100, histtype='bar', color='skyblue', edgecolor='white')
plt.xlim(xmin=-5, xmax=5)
plt.savefig('hist.png')
plt.show
```

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12,6))
values2 = np.linspace(min(values)-10, max(values)+10, 100)
plt.plot(values2, density(values2), color='orange')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='orange')
plt.xlim(xmin=-5, xmax=5)
plt.show
```

```python
plt.subplots(figsize=(12,6))
plt.hist(values, bins=100, histtype='bar', alpha=0.8, density=1, color='skyblue', edgecolor='white')
plt.plot(values2, density(values2), color='orange')
plt.xlim(xmin=-5, xmax=5)
```
